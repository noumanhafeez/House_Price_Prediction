### Data Pre-Processing
#### Data preprocessing is the process of transforming raw data into a usable format for analysis. It's a crucial initial step in data 
#### mining and machine learning tasks. Raw data, whether collected from sensors, surveys, or web scraping, is often messy and unfit for 
#### direct analysis. Here's what data preprocessing aims to achieve:
#### 1. Improve Data Quality: Real-world data can have inconsistencies, errors, missing values, and irrelevant information. Preprocessing 
#### techniques address these issues by cleaning, correcting, and potentially removing outliers or extraneous data points.
#### 2. Transform Data for Analysis: Data may need formatting or conversion to make it suitable for specific algorithms. This might involve 
#### scaling numerical features, encoding categorical variables, or handling dates and times in a consistent way.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
# Set option to display max columns i.e (Display all the columns of the dataframe).
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('House Pricing Dataset/train.csv')
dataset.head()
```

### Let's Handle Missing values

```python
# Let's first deal with categorical nan values.
categorical_feature_nan = [i for i in dataset.columns if dataset[i].isnull().sum()>1 and dataset[i].dtype == 'O']
for i in categorical_feature_nan:
    print(i, dataset[i].isnull().sum() )
```

```python
# Let's fill missing values with 'Missing' a new label. We can get other names by meeting to stakeholders.
dataset[categorical_feature_nan] = dataset[categorical_feature_nan].fillna('NA')
dataset.head()
dataset[categorical_feature_nan].isnull().sum()
```

```python
# Now, deal with numerical missing values
numerical_feature_nan = [i for i in dataset.columns if dataset[i].isnull().sum()>1 and dataset[i].dtype != 'O']
for i in numerical_feature_nan:
    print(i, dataset[i].isnull().sum())
```

```python
# Now, fill missing values with median coz data has many outliers so filling with is not best for dataset.
for i in numerical_feature_nan:
    median_value = dataset[i].median()
    dataset[i].fillna(median_value, inplace=True)
dataset[numerical_feature_nan].isnull().sum()
```

### Now, Let's do scaling features.

```python
scaling_features = [i for i in dataset.columns if i not in ['Id', 'SalePrice']]
scaling_features
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(dataset[scaling_features])
```

```python
scaled_data = scaler.transform(dataset[scaling_features])
scaled_data
```

```python
data = pd.DataFrame(scaled_data, columns=scaling_features)
```

```python
data.head()
```

```python
final_data = pd.concat([dataset[['Id','SalePrice']], data], axis=1)
```
