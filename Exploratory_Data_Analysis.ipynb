### and uncover any potential issues.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
# Set option to display max columns i.e (Display all the columns of the dataframe).
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('House Pricing Dataset/train.csv')
# print shape of dataset with rows and columns
print(dataset.shape)
```

```python
dataset.head()
```

```python
# Let's describe the dataset
dataset.describe()
```

```python
# Let's get info of dataset
dataset.info()
```

### Conclusion:
#### There are many missing values in certain features of dataset as you can see in info of dataset. Let's simplify this in coming cell.

```python
## Let's create list of features which has missing values.
missing_values_features = [i for i in dataset.columns if dataset[i].isnull().sum()>1]
missing_values_features
```

```python
## ohkey, this features has missing values. Now, let's check how many values are missing regarding to feature.
for i in missing_values_features:
    print(i, np.round(dataset[i].isnull().mean(), 3), '% are missing values') 
```

### Conclusion
#### We conclude that the features like [alley, poolqc, miscfeature, fence] has more missing values i.e: approx to 99%. 
#### There are many missing values. So, before we do fill that with some values, let's check their relation with saleprice.

```python
for i in missing_values_features:
    data = dataset.copy()
    
    # 1 indicates for missing values and 0 indicates for non missing values.
    data[i] = np.where(dataset[i].isnull(), 1, 0)
    
    # let's plot it.
    data.groupby(i)['SalePrice'].mean().plot.bar()
    plt.title(i)
    plt.show()
```

### Conclusion:
#### We conclude that some features's missing values are important to saleprice. They can play a significant role regarding to saleprice. 

### Let's check numerical variables and categorical variables.

```python
numerical_features = [i for i in dataset.columns if dataset[i].dtypes != 'O']
categorical_features = [j for j in dataset.columns if dataset[j].dtypes == 'O']
print('The numerical features are:' , len(numerical_features), '\nThe categorical features are:', len(categorical_features))
dataset[numerical_features].head()
```

### There are two types of numerical variables.
#### 1. Discrete variable
#### 2. Continous variable

```python
# Let's check Discrete variable first
discrete_variables = [i for i in numerical_features if len(dataset[i].unique())<20 and i not in ['Id']]
dataset[discrete_variables].head()
```

```python
## Now, find relationship between discrete variables with sale price
for i in discrete_variables:
    data = dataset.copy()
    data.groupby(i)['SalePrice'].mean().plot.bar()
    plt.xlabel(i)
    plt.ylabel('SalePrice')
    plt.title(i)
    plt.show()
```

### Conclusion
#### We can see relationship of different features with SalePrice.

```python
# Now, let's play with continue variables.
continous_feature = [i for i in numerical_features if i not in discrete_variables + ['Id']]
dataset[continous_feature].head()
```

```python
for i in continous_feature:
    data = dataset.copy()
    data[i].hist(bins=20)
    plt.ylabel(i)
    plt.xlabel("count")
    plt.title(i)
    plt.show()
```

### As we see, the graph is not forming bell curve, it's means the data is skewed. We can make it to bell curve by 
### taking logarithmic method.

```python
for i in continous_feature:
    data=dataset.copy()
    if 0 in data[i].unique():
        pass
    else:
        data[i]=np.log(data[i])
        data[i].hist(bins=20)
        plt.xlabel(i)
        plt.ylabel('SalesPrice')
        plt.title(i)
        plt.show()
```

### We notice that data has many outliers. We can check outliers by box plotting.

```python
for i in continous_feature:
    data=dataset.copy()
    if 0 in data[i].unique():
        pass
    else:
        data[i]=np.log(data[i])
        data.boxplot(column=i)
        plt.xlabel(i)
        plt.title(i)
        plt.show()
```

### Now, let's find relationship between categorical and saleprice feature.

```python
for i in categorical_features:
    data = dataset.copy()
    data.groupby(i)['SalePrice'].mean().plot.bar()
    plt.xlabel(i)
    plt.ylabel('SalePrice')
    plt.title(i)
    plt.show()
```

#### As we can see from bar graph for categorical features. We can conclude from this, which category is major in number i.e: from MSZoning, 
#### the category 'FV' is major in number. As the number of FV increasing, the saleprice is also increasing.
