```python
# Let's collect categorical features with NaN.
categorical_features_NA = [i for i in dataset.columns if dataset[i].isnull().sum()>0 and dataset[i].dtype == 'O']
categorical_features_NA
```

```python
# fill Nan with NA, which has meaning. You can have a look at description of this dataset.
dataset[categorical_features_NA] = dataset[categorical_features_NA].fillna('NA')
dataset[categorical_features_NA].isnull().sum()
```

```python
dataset.info()
```

### Let's check numerical variables and categorical variables.

```python
numerical_features = [i for i in dataset.columns if dataset[i].dtypes != 'O']
categorical_features = [j for j in dataset.columns if dataset[j].dtypes == 'O']
print('The numerical features are:' , len(numerical_features), '\nThe categorical features are:', len(categorical_features))
dataset[numerical_features].head()
```

### There are two types of numerical variables.
#### 1. Discrete variable
#### 2. Continous variable

```python
# Let's check Discrete variable first
discrete_variables = [i for i in numerical_features if len(dataset[i].unique())<20 and i not in ['Id']]
dataset[discrete_variables].head()
```

```python
# Let's find distribution plot of saleprice
sns.distplot(dataset['SalePrice'], kde=False)
```

```python
# You see, the graph of SalePrice is right skeewed.
dataset['SalePrice'].describe()
```

```python
#Let's see the correlation of numerical features
correlation = dataset[numerical_features].corr()
print(correlation['SalePrice'].sort_values(ascending=False), '\n')
```

```python
sns.heatmap(correlation, square=True, vmax=0.8)
```

```python
# set to 10 features and show their relations.
k = 10
cols = correlation.nlargest(k,'SalePrice')['SalePrice'].index
print(cols)
cm = np.corrcoef(dataset[cols].values.T)
f, ax = plt.subplots(figsize=(14,12))
sns.heatmap(cm, vmax=0.6, square=True, linewidths=0.02, annot=True, xticklabels=cols.values, yticklabels=cols.values)
```

### We see that garageCar and garageArea are highly correlated to each i.e they've high multicollinearity. It can issue for model.

```python
## Now, find relationship between discrete variables with sale price
for i in discrete_variables:
    data = dataset.copy()
    data.groupby(i)['SalePrice'].median().plot.bar()
    plt.xlabel(i)
    plt.ylabel('SalePrice')
    plt.title(i)
    plt.show()
```

```python
## Now, let's use scatter plotting for better visualization.
for i in discrete_variables:
    data = dataset.copy()
    plt.scatter(data[i], dataset['SalePrice'])
    plt.xlabel(i)
    plt.ylabel('SalePrice')
    plt.title(i)
    plt.show()
```

```python
# Let's see trend between GrLivArea with SalePrice with regplot.
sns.regplot(x='GrLivArea', y='SalePrice', data=dataset, scatter=True, fit_reg=True)
```

### Conclusion
#### We can see relationship of different features with SalePrice.

```python
# Now, let's play with continue variables.
continous_feature = [i for i in numerical_features if i not in discrete_variables + ['Id']]
dataset[continous_feature].head()
```

```python
for i in continous_feature:
    data = dataset.copy()
    data[i].hist(bins=20)
    plt.ylabel(i)
    plt.xlabel("count")
    plt.title(i)
    plt.show()
```

### As we see, the graph is not forming bell curve, it's means the data is skewed. We can make it to bell curve by 
### taking logarithmic method.

```python
for i in continous_feature:
    data=dataset.copy()
    if 0 in data[i].unique():
        pass
    else:
        data[i]=np.log(data[i])
        data[i].hist(bins=20)
        plt.xlabel(i)
        plt.ylabel('SalesPrice')
        plt.title(i)
        plt.show()
```

### We notice that data has many outliers. We can check outliers by box plotting.

```python
for i in continous_feature:
    data=dataset.copy()
    if 0 in data[i].unique():
        pass
    else:
        data[i]=np.log(data[i])
        data.boxplot(column=i)
        plt.xlabel(i)
        plt.title(i)
        plt.show()
```

### Now, let's find relationship between categorical and saleprice feature.

```python
for i in categorical_features:
    data = dataset.copy()
    data.groupby(i)['SalePrice'].mean().plot.bar()
    plt.xlabel(i)
    plt.ylabel('SalePrice')
    plt.title(i)
    plt.show()
```

#### As we can see from bar graph for categorical features. We can conclude from this, which category is major in number i.e: from MSZoning, 
#### the category 'FV' is major in number. As the number of FV increasing, the saleprice is also increasing.
