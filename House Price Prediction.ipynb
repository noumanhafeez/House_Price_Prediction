### 3.Data Pre-processing steps

```python
# So the data in hist portion says that it is skeewed data. So, they need to be convert into bell curve distribution
trained_data['total_rooms'] = np.log(trained_data['total_rooms']+1)
trained_data['total_bedrooms'] = np.log(trained_data['total_bedrooms']+1)
trained_data['population'] = np.log(trained_data['population']+1)
trained_data['households'] = np.log(trained_data['households']+1)
```

```python
trained_data.hist(figsize=(15,8))
```

```python
# Find categorical col and then convert to nummerical col
trained_data.ocean_proximity.value_counts()
```

```python
# Use one hot encoding to convert categorical to nummerical
trained_data = trained_data.join(pd.get_dummies(trained_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
```

```python
trained_data
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(trained_data.corr(), annot=True, cmap='YlGnBu')
```

```python
plt.figure(figsize=(15,9))
sns.scatterplot(x='latitude' , y='longitude' , data=trained_data, hue='median_house_value', palette='coolwarm')
```

### 4.Feature Engineering steps

```python
trained_data['bedroom_ratio'] = trained_data['total_bedrooms'] / trained_data['total_rooms']
trained_data['households_rooms'] = trained_data['total_rooms'] / trained_data['households']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(trained_data.corr(), annot=True, cmap="YlGnBu")
```
